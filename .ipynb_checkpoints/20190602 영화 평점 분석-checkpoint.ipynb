### 2. 인기 있는 영화 10개 찾기

- 기준 : 평균 평점이 4.5 이상이고 평점 개수가 2000개 이상인 영화

```python
# 영화별로 평점의 평균, 개수
# 로우 : 제목 / 영화아이디
# 값 : 평점
# 함수 : 평균, 개수
tmp=df.pivot_table(index='제목',values='평점',aggfunc=['mean','count']).swaplevel(1,0,axis=1)
```

```python
tmp.head(3)
```

```python
# 평균평점 : 4.5 이상
# &
# 평점개수 : 2000개 이상
tmp[(tmp[('평점','mean')] >= 4.5) & (tmp[('평점','count')]>=2000)]
```

```python
# 10개 채우기
# 1. 평점이 높은 순서로 상위 10
tmp[(tmp[('평점','mean')] >= 4) & (tmp[('평점','count')]>=2000)].sort_values(('평점','mean'),ascending=False).head(10)
```

```python
s1=df.groupby('제목')['평점'].mean()
s1
```

```python
df.head(2)
```

```python
s2=s1.reset_index().copy()
```
