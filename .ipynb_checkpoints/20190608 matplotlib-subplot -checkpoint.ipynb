# Matplotlib 객체 구조
- Figure : 한 개 이상의 Axes 객체 포함하는 가장 큰 단위의 객체(캔버스 틀)
- Axes : 두 개 이상의 Axis 객체 포함(캔버스 안의 하위 영역)
- Axis: 축 정보를 담고 있는 객체

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 한글 폰트 설정
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

#### 1. Figure 생성
- plt.figure() 함수 : figure 객체 반환
    - plt.plot() 함수를 실행하면 자동으로 figure 객체를 생성
    - figure 객체의 크기(plot 이미지의 크기)를 설정, 생성하고 싶을때 명시적으로 사용

```python
# figure() 함수를 사용하지 않고 plot()으로 생성되는 figure 객체
data=np.random.randn(100)
plt.plot(data)
```

```python
# figure 객체 크기 생성
# plt.figure(figsize=(w,h))
plt.figure(figsize=(10,1))
plt.plot(data)
```

#### 2. Subplot 생성
- 하나의 윈도우안에 여러개의 이미지를 생성해야하는 경우
- subplot() 함수
    - plt.subplot(row,col,idx)
        - 단독 생성: idx 값으로 axes의 위치 지정
        - 2,2,1=>221
    - plt.subplots(row,col)
        - figure,axes객체 생성: 행/열구조만 설정하여 axes 객체 생성한 후 각각의 axes객체로 위치 지정

```python
# 1. subplot()으로 단독 생성
# 2행1열짜리 레이아웃 생성
# 첫번쨰 플롯
plt.subplot(2,1,1)
```

```python
# 한번에 모든 subplot 화면에 띄우기
# 현재는 틀만 그린 상태
plt.subplot(2,1,1)
plt.subplot(2,1,2) # 두개를 한 셀에 써야됨
```

```python
# subplot 단독 생성
# 구조 2*2 => 4칸 짜리 레이아웃 => 서로 다른 모양의 라인 플롯
# 데이터 : random.rand(5)=> 0~1 사이 난수 5개
# 순서
# 1. 틀 생성
# 2. 값을 그래프로 그리기
# 3. 속성
# 4. show()
plt.subplot(2,2,1)
plt.plot(np.random.rand(5))
plt.title('axes1')
plt.subplot(2,2,2)
plt.plot(np.random.rand(5))
plt.title('axes2')
plt.subplot(2,2,3)
plt.plot(np.random.rand(5))
plt.title('axes3')
plt.subplot(2,2,4)
plt.plot(np.random.rand(5))
plt.title('axes4')
plt.tight_layout(pad=3)
plt.show()
```

```python
# figure 객체와 여러 개의 axes 객체를 생성하는 방법
# plt.subplots(row,col)
# plt.subplot(2,2,1) => 위치지정하는 법
# axes[0,0],[0,1],[1,0],[1,1]
fig,axes=plt.subplots(2,2)
axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title('axes1')
axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title('axes2')
axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title('axes3')
axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title('axes4')

plt.tight_layout(pad=3)
```

#### 3. x축을 공유하면서 서로 다른 y축을 가지는 plot 생성
- axes 객체를 생성하여 서로 다른 y축 적용
    - axes.twinx(): 동일한 x축을 공유하는 axes를 반환

```python
plt.plot([10,5,6,17,20],'r-')
plt.plot([1000,3000,10000,7000,15000],'g:')
```

```python
# 데이터 : 빨간실선, 녹색점선으로 구분해서 라인플롯
# plt.subplots()
fig,ax0 = plt.subplots()
ax1=ax0.twinx()

ax0.plot([10,5,6,17,20],'r-')
# 공통 속성(이미지 타이틀, x축 이름): 원본객체에서 속성 반영해야함
ax0.set_title('공통 x축을 가지는 서로 다른 y축')
ax0.set_ylabel('y0')
ax0.set_xlabel('x')
ax1.plot([1000,3000,10000,7000,15000],'g:')
ax1.set_ylabel('y1')
# ax1.set_xlabel('x')   # 안됨
# plt.xlabel('x') # 안됨
plt.show()
```

## 연습문제

```python
# 데이터 생성
# x축으로 사용할 데이터 : 범주형 데이터(2~3)
# y축으로 사용할 데이터 : 정수값 데이터 => 각각의 범주가 가지는 크기
# 세반의 평균 점수
class_name=['A','B','C']
scores=[70,85,90]
```

```python
# 크기가 (4,3)인 캔버스(figure) 객체 생성
# bar plot
plt.figure(figsize=(4,3))
plt.bar(class_name,scores)
```

```python
# scatter plot
plt.figure(figsize=(4,3))
plt.scatter(class_name,scores)
```

```python
# line plot
plt.figure(figsize=(4,3))
plt.plot(class_name,scores)
```

```python
# 모든 subplot을 한번에 보기(1*3)
# 1번위치 : bar
# 2번위치 : scatter
# 3번 위치 : line
# plt.subplot(r,c,idx)
plt.subplot(1,3,1)
plt.bar(class_name,scores)
plt.subplot(1,3,2)
plt.scatter(class_name,scores)
plt.subplot(1,3,3)
plt.plot(class_name,scores)
plt.tight_layout(pad=3)
```

```python
# subplot 3*1
# axes 객체 사용
# plt.subplots(r,c)
# row 나 col 이 1일때는 인덱스 번호 하나만 
fig,axes = plt.subplots(3,1,figsize=(3,5)) # 그림 크기
axes[0].bar(class_name,scores)
axes[1].scatter(class_name,scores)
axes[2].plot(class_name,scores)
plt.tight_layout(pad=3) # 여백
```
