### 2. 원본 데이터를 활용하여 종이름 데이터를 컬럼으로 추가
 - 최종적으로 저장되는 값은 숫자로 구분된 데이터

```python
dataset.keys()
```

```python
dataset['target_names']
```

```python
df['Species']=dataset['target']
```

```python
df.head()
```

```python
dataset['target_names']
```

```python
# 리스트를 이용한 추가
tmp=[]
for i in dataset['target']:
    if i==0:
        tmp.append('setosa')
    elif i ==1:
        tmp.append('versicolor')
    else:
        tmp.append('virginica')
```

```python
df['species']=tmp
```

```python
df.head()
```

```python
df.tail()
```

```python
# 딕셔너리 타입을 이용해서 맵핑을 수행(조건문을 생랼할 수 있는 방법)
# 딕셔너리 타입 : 종번호 & 종이름
dict_map={0:'setosa',1:'versicolor',2:'virginica'}
# zip() 함수, dict() 함수
# zip(x,y)
dict_map2=dict(zip(range(3),dataset['target_names']))
# a=4 if x>2 else 10
df['species']= [dict_map[i] for i in dataset['target']]
```

## Pie plot(원형차트)
- 전체에서 특정 범주가 차지하는 비중/비율을 파악하기 위해 사용
- pandas.DataFrame/Series.plot(kind='pie')
    - Series 객체 : 하나의 원형 차트 생성
    - DataFrame 객체: rkr duf akek gkskTlr dnjsgud ckxm todtjd
- pandas.DataFrame/Series.plot.pie(y=None,figsize=None,subplots=False)
    - y: 하나의 컬럼에 대해 원형 차트를 생성할 때 사용할 컬럼명
    - figsize : 생성할 figure 객체 크기(tuple)
    - subplots: y값으로 하나의 컬럼을 사용하는 것이 아니라 전체 df를 사용할 때 True 로 지정

```python
# 전체 종별 데이터 평균
df1=df.groupby('species').mean()
df1
```

```python
# 하나의 속성별로 종별 파이 차트 생성
# y_label
df1.plot(kind='pie',subplots=True,layout=(2,2),legend=False,label='')
plt.show()
```

```python
# 추가 파라미터 : label(컬럼이름),title
df1.plot(kind='pie',y='sepal length (cm)',label='',title='<sepal length>',legend=False)
plt.show()
```

## scatter-plot(산포도)
- 두 종류의 변수만 포함하여 적용할 수 있음
- 두 변수 간의 관계에 대해 분산 분포를 통해 파악할 수 있음
- pandas.DataFrame/Sereies.plot(kind='scatter',x=None,y=None)
    - x: 가로축에 사용될 데이터/컬럼명
    - y: 세로축에 사용될 데이터/컬럼명

```python
# 모든 종에 대하여 산포도
df.plot(kind='scatter',x='petal length (cm)',y='petal width (cm)')
```

```python
# 종별로 scatter plot
for s in df['species'].unique():    # dataset['target_names']
    df[df['species'] == s].plot(kind='scatter',x='petal length (cm)',y='petal width (cm)')
    plt.title(s)
```

```python
# 종 기준으로 산포도 색상을 다르게 저장
# df.plot.scatter(x,y,c)
# color 적용을 위한 정수형 종구문을 컬럼으로 추가
# dataset['target']
df['map']=dataset['target']
```

```python
df.head(1)
```

```python
df.plot.scatter(x='petal length (cm)',y='petal width (cm)',c='map',colormap='plasma')
```
